```python
from FemalePreds import *

# ---------------------
# 4) Main execution
# ---------------------

if __name__ == "__main__":
    # 1) Read data
    AM = pd.read_csv('../Step1_LoadingAndCleaningData/AM.csv')
    FM = pd.read_csv('../Step1_LoadingAndCleaningData/FM.csv')
    AP = pd.read_csv('../Step1_LoadingAndCleaningData/AP.csv')
    FP = pd.read_csv('../Step1_LoadingAndCleaningData/FP.csv')
    AMP = pd.read_csv('../Step1_LoadingAndCleaningData/AMP.csv')
    FMP = pd.read_csv('../Step1_LoadingAndCleaningData/FMP.csv')
    
    datasets = {
        'FM': FM,
        'AM': AM,
        'AP': AP,
        'FP': FP,
        'FMP': FMP,
        'AMP': AMP
    }

    # 2) Define train/test groups
    train_groups = ["M_Rapa", "M_Cana", "M_17aE2", "M_CR", "M_Aca", "M_Cont_12"]
    predict_groups = ["F_Rapa", "F_Cana", "F_17aE2", "F_CR", "F_Aca", "F_Cont_12"]
    
    # 3) We'll run multiple iterations (set how many you want here)
    n_iterations = 20
    
    all_predictions = pd.DataFrame()
    
    # 4) Iterate over each dataset and get median predictions
    for dataset_name, df in datasets.items():
        print(f"\n\nProcessing dataset: {dataset_name}")
        try:
            preds = train_and_predict_on_full_data(
                df, 
                dataset_name, 
                train_groups, 
                predict_groups, 
                n_iterations=n_iterations
            )
            if not preds.empty:
                all_predictions = pd.concat([all_predictions, preds], ignore_index=True)
            else:
                print(f"No predictions for dataset: {dataset_name}")
        except Exception as e:
            print(f"Error processing dataset {dataset_name}: {e}")
    
    # 5) Save final median predictions if any
    if not all_predictions.empty:
        output_name = "012825_Predictions_on_Females_Median_of_20Iterations.xlsx"
        all_predictions.to_excel(
            output_name, 
            index=False,
            columns=['Dataset', 'Whole_Training_Set', 'Test_Group', 'Mouse', 'Prediction']
        )
        print(f"Predictions saved to {output_name}")
    else:
        print("No predictions were made.")
```
